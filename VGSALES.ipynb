```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('vgsales.csv')
df
```

```python
df.dtypes
```

```python
df.columns
```

```python
# Create
new_row = {
    'Rank': 16601,
    'Name': 'New Game',
    'Platform': 'PC',
    'Year': 2025,
    'Genre': 'Action',
    'Publisher': 'New Publisher',
    'NA_Sales': 0.50,
    'EU_Sales': 0.30,
    'JP_Sales': 0.00,
    'Other_Sales': 0.20,
    'Global_Sales': 1.00
}

df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
df
```

```python
#Read
action_games = df[df['Genre'] == 'Action']
print(action_games)
```

```python
#update
df.loc[df['Name'] == 'New Game', 'Global_Sales'] = 1.5
df
```

```python
# delete 
df = df[df['Publisher'] != 'New Publisher']
```

```python
top_games = df.sort_values(by='Global_Sales', ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(data=top_games, x='Global_Sales', y='Name', palette='viridis')
plt.title('Top 10 Best Selling Games Globally')
plt.xlabel('Global Sales (Millions)')
plt.ylabel('Game Name')
plt.tight_layout()
plt.show()
```

```python
sales_region = df.groupby('Genre')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()

plt.figure(figsize=(10,6))
sns.heatmap(sales_region, annot=True, fmt=".1f", cmap='coolwarm')
plt.title('Sales by Region per Genre')
plt.xlabel('Region')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```

```python
top_publishers = df['Publisher'].value_counts().head(10).index
df_top_publishers = df[df['Publisher'].isin(top_publishers)]

plt.figure(figsize=(14,6))
sns.boxplot(data=df_top_publishers, x='Publisher', y='Global_Sales')
plt.title('Sales Distribution by Publisher')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
games_per_year = df['Year'].value_counts().sort_index()

plt.figure(figsize=(12,6))
sns.lineplot(x=games_per_year.index, y=games_per_year.values)
plt.title('Number of Games Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Games')
plt.tight_layout()
plt.show()
```
